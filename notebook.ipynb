# Image Processing Basics in Python

In this notebook, we will explore some basic operations with images using Python libraries such as OpenCV and Matplotlib. We will go through the following tasks:

- Import necessary libraries
- Load an image
- Display image properties
- Convert image formats
- Display the image and its color channels

```python
# Task 1
# Import the required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Confirm that the imports worked
print("Libraries imported successfully!")
```

## Loading and Displaying an Image

We'll first identify the image we want to load, then read it into memory using OpenCV.

```python
# Task 2
# Load the image from the pre-seeded file
file_name = 'sample.jpg'

# Read the image using cv2.imread
image_bgr = cv2.imread(file_name)

# Verify the image has loaded
if image_bgr is not None:
    print(f"Image '{file_name}' loaded successfully!")
else:
    print(f"Failed to load image '{file_name}'.")
```

## Image Properties

We will inspect some basic properties of the image, such as its type and dimensions.

```python
# Task 3
# Check the image type and shape
image_type = type(image_bgr)
image_shape = image_bgr.shape

print(f"Type of image: {image_type}")
print(f"Shape of image: {image_shape}")
```

## Converting the Image Format

We will convert the image from BGR (default in OpenCV) to RGB format for better visualization with Matplotlib.

```python
# Task 4
# Convert image_bgr to RGB
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Confirm the conversion
print("Image converted from BGR to RGB.")
```

## Displaying the Image

We will use Matplotlib to display the converted RGB image.

```python
# Task 5
# Display the RGB image using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Turn off the axes
plt.title('RGB Image')
plt.show()
```

## Extracting Color Channels

Finally, we'll extract and display the individual Red, Green, and Blue color channels of the image.

```python
# Task 6
# Extract Red, Green, and Blue channels
red_channel = image_rgb[:,:,0]
green_channel = image_rgb[:,:,1]
blue_channel = image_rgb[:,:,2]

# Display each channel
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(red_channel, cmap='Reds')
ax[0].set_title('Red Channel')
ax[0].axis('off')

ax[1].imshow(green_channel, cmap='Greens')
ax[1].set_title('Green Channel')
ax[1].axis('off')

ax[2].imshow(blue_channel, cmap='Blues')
ax[2].set_title('Blue Channel')
ax[2].axis('off')

plt.show()
```
